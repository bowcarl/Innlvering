### Oppgave 1

a.) En sekant er en relllinjet linje som går igjennom to punkter på en funkjsjon, fordi k-konstanten skjærer x og x + $\Delta x$ i funksjonen f

b.) En tagent er et spesifikt punkt sitt stigningstall på en funksjon. stigningstallet k til sekanten gjennom punktet a og delt_a på funksjonen f kan skrives som

$$k = \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

vi vet at den deriverte av tangeten er stigningstallet utrykkt som en konstant k

c.) Derivasjon er oppfattningen av endring mellom to punkter

d.) Forskjellen mellom Newtons kvotient og definisjonen av den deriverte er at, kvotienten av den deriverte er en numerisk måte er regne derivasjon den er mye mer presis en analytisk altså definisjonen av den deriverte, Newton som forsket innenfor fysikken synes det var tidskrevende å gjøre titals tusener av utregninger for hver gang han skulle finne et stigningstall når han brukte den numeriske metoden, han oppfant derfor den analystiske derivasjonen hvor man antar at $\frac{\lim}{\Delta x\to 0}$ dette viste seg å gjøre problemet mye lettere. Det er denne teknikken som blir brukt i dagen matematikk. Her sier vi at $\Delta x$ blir tilsvarende lik $0$ og se bort i fra verdien.
<center> Den deriverte kan skriver slik <center>

$$f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

```python
from pylab import *

def fder(f, x, dx):
    return (f(x+dx)-f(x))/dx

def f(x):
    return x**2+23*x+3

analytisk_verdi = 25


dx = []
feilen = []

for i in range(1,17):
    DX = 10**(i*(-1))
    dx.append(log10(DX))
    FEILEN = abs(analytisk_verdi-fder(f, 1, DX))
    feilen.append(log10(FEILEN))

plot(dx, feilen)
ylabel("Feilmargin i logarytmisk skala")
xlabel("d(x) i logarytmisk skala")
grid()
show()
```

## Derivasjon av Heistur

```python
from pylab import *

psn_t = loadtxt("Pressureandvelocity.csv", skiprows = 1, delimiter = "," , usecols = (0, 2))

t = psn_t[:, 0]
s = psn_t[:, 1]

plot(t, s, label = "$s$")
grid()
title("Heistur")
xlabel("Tiden")
ylabel("Strekningen")
legend()
show()

v = []
v.append(0)

for i in range(len(s)-1):
    sder = (s[i+1]-s[i])/(t[i+1]-t[i])
    v.append(sder)

plot (t, v, color = "red", label = "$V$")
grid()
xlabel("Tiden")
ylabel("Farten")
legend()
show()

a = []
a.append(0)

for j in range(len(v)-1):
    vder = (v[j+1]-v[j])/(t[j+1]-t[j])
    a.append(vder)

plot(t, a, color = "green" , label = "$a$")
grid()
xlabel("Tiden")
ylabel("Akselerasjonen")
legend()
show()
```
